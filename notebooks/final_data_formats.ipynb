# How data tables should look like after each preprocessing step

```python
import pandas as pd
```

## Raw Data

```python
raw_data = pd.read_csv('../raw_data/xy_hc_fai_tstp_E2_2_PM.csv')
raw_data.head()
```

## Preprocess 

```python
preproc = pd.read_csv('../raw_data/example_data_format/preprocess.txt')
preproc = preproc.drop(columns=['Unnamed: 0','Unnamed: 1'])
preproc
```

## Data Summary Table

```python
summary = pd.read_csv('../raw_data/example_data_format/data_summary.txt')
summary = summary.drop(columns=['Unnamed: 1'])
summary = summary.set_index('E2_C100_DataSUMMARY').rename(columns={'Unnamed: 2': ''})
summary.head()
```

## Event List

```python
events = pd.read_csv('../raw_data/example_data_format/event_list.txt')
events = events.drop(index=0,columns=['Unnamed: 0','Unnamed: 1']).set_index('Event_n')
events.head()
```

## Fish List

```python
fish_list = pd.read_csv('../raw_data/example_data_format/fish_list.txt')
fish_list = fish_list.drop(index=0, columns=['Unnamed: 0','Unnamed: 1']).set_index('fish_n')
fish_list
```
